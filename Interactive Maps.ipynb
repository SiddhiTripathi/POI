```python
# Importing libraries
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
df = pd.read_csv("../POI.csv")
```

```python
# Map of commercial buildings in Mumbai

# Mumbai coordinates
DF_COORDINATES = (19.0760, 72.8777)
 
MAX_RECORDS = 240
  
# Create empty map zoomed in on Mumbai
map = folium.Map(location=DF_COORDINATES, zoom_start=12)
 
# Add a marker for every record in the filtered data
for each in df[0:MAX_RECORDS].iterrows():
    folium.Marker(location = [each[1]['Y'],each[1]['X']]).add_to(map)

display(map)
map.save('commercial.html')
```

```python
# Map of clusters of commercial buildings in Mumbai

# Mumbai Coordinates
DF_COORDINATES = (19.0760, 72.8777)
 
MAX_RECORDS = 240
  
# create empty map zoomed in on Mumbai
map = folium.Map(location=DF_COORDINATES, zoom_start=12)

mc= MarkerCluster()
 
# add a marker for every record in the filtered data, use a clustered view
for each in df[0:MAX_RECORDS].iterrows():
    mc.add_child(folium.Marker(location = [each[1]['Y'],each[1]['X']], clustered_marker = True))
    
map.add_child(mc)
    
display(map)
map.save('clusters.html')
```

```python
# The map shows that South Mumbai has the highest density of commercial buildings, and it is labelled as 'Central Business District'.
# Moving up to areas like Lower Parel, Mahalaxmi, Worli are commercially dense enough and called as 'Extended Central Business District'.
# In central regions near airport such as Kalina, BKC are known as 'Emerging Central Business District'; and Andheri, Powai, Kurla etc are called as
# 'Secondary Business District'.These localities comprise the largest number of commercial points in Mumbai, around greater than 50.
# And the other areas like Thane, Navi Mumbai are also decent commercial areas, classified as 'Peripheral Business District'.
```
